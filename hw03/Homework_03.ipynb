## 2. What is wrong with the following code and how would you fix it?
```C++
int* pointerToMax(int a, int b) {
    if(a>b) return &a;
    else    return &b;
}
```
- Problem is that this method is inputting int itself, not a pointer to the int. Thefore I'd import pointer to the int a and b instead, and change it to..

```C++
int* pointerToMax(int* a, int* b){
    if(*a>*b) return a;
    else      return b;
}
```


## 3. What's wrong with the following code and how would you fix it?
```C++
int* ptr = NULL;
scanf("%d", ptr);
printf("You entered %d", &ptr)
```
- There are several problems with the code.
1. `ptr` is not initalized to point any variables, therefore it can be pointing at random locations in the memory, possibly causing it to raise segmentation fault. 
2. No print statement before the `scanf`, can be confusing to users of the program
3. `;` is missing on the end of the last line as well as `&` is used to reference? the value in `ptr`.

Solution would be..
```C++
int usr_in; // create var that ptr points to.
int* ptr = &usr_in; // initialize the ptr.
scanf("%d", ptr);
printf("You entered %d", usr_in); // or printf("You entered %d", *ptr);
```

or...
```C++
int usr_in;
printf("enter an int: ");
scanf("%d", &usr_in);
printf("You've entered %d", usr_in);
```

## 4. Describe what each of the following declare:
1. `int* a;`: pointer `a` to int value
2. `const int b;`: constant integer named b
3. `const int* c;`: pointer to const int, pointer is changable, but not the int that is pointing at
4. `int* const d;`: const pointer that points at integer. the pointer d can't be changed, but value in the int can be modified.
5. `const int* const e;`: constant pointer e, that points to a const integer. Neither pointer or value in the const int are changable.

## 5. What is the difference between the * and & operators??
- `*`: is dereferencing operator, for example
> ```C++
     int* ptr; // this initialize a pointer
     *ptr; // this is the value that ptr is pointing at.
  ```

- `&`: is referencing operator, for example
> ```C++
     int var; //create a variable
     &var; // this is a pointer that points at var.
```

## 6. What is the difference between const int* a and int* const b??

- `const int* a`: is pointer to constant integer.
- `int* const b`: is a constant pointer that points to an int. 

Difference here is that one being constant pointer vs the other being pointer, which points at constant int.
