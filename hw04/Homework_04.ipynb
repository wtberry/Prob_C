### 1. What is wrong with the following code, and how would you fix it?

```C
void swap(int* a, int* b) {
    int* tmp = a;
    a = b;
    b = tmp;
} // end of swap method
```
_This method only modifies the pointers of the numbers, which won't have effect outside of the methods. This can be fixed as _
```C
void swap(int* a, int*b){
    int temp;
    temp = *a; //value of a
    a* = *b; //value of b
    b* = temp // value of a in b
}
```

### 2. What is wrong with the following code, and how would you fix it?
```C
char copy[8];
const char* string = "hello";
for(int i=0; string[i] != ’\0’; ++i) {
copy[i] = string[i];
}
```
_One thing could be the size of `copy` array is too big. And the for loop won't go all the way til the `\0` of `string`. So it could be written as..._

```C
char copy[6]; //hello + \0
const char* string = "hello";
int i = 0;
for( ; string[i] != `\0`; ++i){
    copy[i] = string[i];
}
copy[i] = string[i]; // copy the \0
```


### 4. What is wrong with the following code, and how would you fix it?
```C
double* allocateArray(int length) {
    double array[length];
    return array;
}
```
_returned array would be out of scope, so needed to be allocated in heap instead._

```C
double* allocateArray(int length){
    double* array = new double[length];
    return array;
}
```
