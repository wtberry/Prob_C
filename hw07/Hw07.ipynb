## 3. Describe each of the following methods.

1. `int* method(int* arg);`
> method to return pointer to int, and input pointer to int. might modify the class.
2. `const int* method(int* arg);`
> returns pointer to constant int, might modify the class, and input is pointer to int arg.

> returns constant pointer to constant int, might modify class and inputs pointer to int arg.
